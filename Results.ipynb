Nas etapas anteriores, fizemos a coleta, tratamento e análise dos *tweets*. Agora, vamos trazer uma análise sobre o sentimento e tópicos de cada *tweet*.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

## Carga de Dados

Esta seção serve para carregarmos em um DataFrame todas os nossos *tweets* formatados e, então, realizar nossas análises

```python
# dados de entrada (pasta e arquivo)
data_input_folder = 'Tweets Analisados'
data_input_filename = '[Analisado] - Resultado FLU x SCO.csv'
```

```python
tweets_df = pd.read_csv(f'{data_input_folder}/{data_input_filename}', sep=',')
```

```python
tweets_df
```

## Análise de Dados

Nesta seção realizaremos a construção de uma nuvem de palavras, onde cada palavra terá seu tamanho definido pelo número de ocorrências e sua cor será de acordo com o sentimento médio de cada uma delas.

```python
# cria um dicionário de tópico: sentimento
topic_sentiment = {}
for sentiment, topic in zip(tweets_df['Sentimento'], tweets_df['Tópico']):
    if topic in topic_sentiment:
        # faz uma média simples 'grosseira' de sentimentos para o mesmo tópico
        topic_sentiment[topic.replace(" ", "_")] = topic_sentiment[topic] + sentiment / 2
    else:
        topic_sentiment[topic.replace(" ", "_")] = sentiment
```

### Estatística Básica

Vamos ver qual porcentagem dos *tweets* são positivas, neutras e negativas.

```python
positive_rate = 0
neutral_rate = 0
negative_rate = 0
total = len(tweets_df)
```

```python
for topic in topic_sentiment:
    if topic_sentiment[topic] > 0:
        positive_rate = positive_rate + 1
    elif topic_sentiment[topic] < 0:
        negative_rate = negative_rate + 1
    elif topic_sentiment[topic] == 0:
        neutral_rate = neutral_rate + 1
```

```python
print(f"A taxa de tweets positivos é de {round(positive_rate * 100  / total, 2)}")
print(f"A taxa de tweets neutros é de {round(neutral_rate * 100  / total, 2)}")
print(f"A taxa de tweets negativos é de {round(negative_rate * 100  / total, 2)}")
```

### Nuvem de Palavras

Agora, vamos criar uma nuvem de palavras com os termos analisados.

```python
def color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    '''
        Função para calcular a cor de cada palavra de acordo com o sentimento associado a ela.
    '''
    value = topic_sentiment.get(word, 0)
    
    if value < 0:
        # assinala tons de vermelho para sentimentos negativos ( < 0)
        red = max(0, min(255, int(255 * (1 + value))))
        green = 0
    else:
        # assinala tons de verde para sentimentos positivos ( >= 0)
        red =  0
        green = max(0, min(255, int(255 * (1 - value))))
    return f"rgb({red}, {green}, 0)"  # string para representar nosso RGB
```

```python
# cria objeto wordcloud
wordcloud_obj = WordCloud(color_func=color_func, background_color='white', max_words=500, width=1200, height=900,)
wordcloud_obj.generate(" ".join(topic_sentiment.keys()))
```

```python
# cria a imagem
plt.figure(figsize=(12, 9))
plt.imshow(wordcloud_obj, interpolation='bilinear')
plt.axis('off')
plt.show()
```
